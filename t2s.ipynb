```python
class TimeoutException(Exception):
    pass
```

```python
class ThreadWithReturn(threading.Thread):
    def __init__(self, target, args=()):
        super().__init__()
        self._target = target
        self._args = args
        self._return = None
        self._exc = None

    def run(self):
        try:
            self._return = self._target(*self._args)
        except Exception as e:
            self._exc = e

    def join(self, timeout=None):
        super().join(timeout)
        if self.is_alive():
            raise TimeoutException("Timed out!")
        if self._exc:
            raise self._exc
        return self._return
```

```python
def clean_page_text(page, prev_header=None):
    blocks = page.get_text("blocks")
    blocks.sort(key=lambda b: b[1])
    page_height = page.rect.height
    filtered_blocks = []

    for b in blocks:
        x0, y0, x1, y1, text, *_ = b
        text = text.strip()
        if not text:
            continue
        if y0 < 50 or y1 > page_height - 50:
            continue
        if re.fullmatch(r"\d+", text):
            continue
        if prev_header and text == prev_header:
            continue
        if re.match(r"^(Figure|Table|Fig\.|Tab\.)\s*\d+", text, re.IGNORECASE):
            continue
        # Drop math-heavy blocks (many symbols, equations)
        if re.search(r"[∑∫√≈±≤≥∞λμνΩπ]", text):
            continue
        filtered_blocks.append(text)

    return " ".join(filtered_blocks)
```

```python
def pdf_to_pages(pdf_path):
    doc = fitz.open(pdf_path)
    pages = []
    prev_header = None
    for page in doc:
        text = clean_page_text(page, prev_header=prev_header)

        blocks = page.get_text("blocks")
        if blocks:
            candidate_header = blocks[0][4].strip()
            if candidate_header and len(candidate_header.split()) < 5:
                prev_header = candidate_header

        text = text.replace("\n", " ")
        text = re.sub(r"(^|\s)(\d{1,2})\s+([A-Z][A-Z\s]+)", r"\1\2.\n\3", text)
        text = re.sub(r"(^|\s)(\d+\.\d+)\s+([A-Z][A-Z\s]+)", r"\1\2.\n\3", text)
        text = re.sub(r"\s+", " ", text).strip()

        pages.append(text)
    return pages
```

```python
def pdf_to_audiobook(pdf_path, out_path, model="tts_models/en/vctk/vits",
                     page_range=None, speaker="p274"):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    tts = TTS(model).to(device)
    pages = pdf_to_pages(pdf_path)

    if page_range:
        start, end = page_range
        pages = pages[start-1:end]  # 1-based input → Python 0-based

    audio_streams = []
    for i, text in enumerate(pages, 1):
        print(f"Rendering page {i}/{len(pages)} on {device}...")

        # Use smaller chunks to avoid GPU OOM
        chunks = [text[j:j+800] for j in range(0, len(text), 800)]
        for chunk in chunks:
            wav = tts.tts(chunk, speaker=speaker)
            audio_streams.append(wav)

    # Concatenate into one long array in correct page order
    full_audio = np.concatenate(audio_streams)
    sf.write(out_path, full_audio, samplerate=tts.synthesizer.output_sample_rate)
    print(f"Saved audiobook: {out_path}")

    # Free memory
    del audio_streams
    torch.cuda.empty_cache()
```

```python
class TimeoutException(Exception):
    pass

class ThreadWithReturn(threading.Thread):
    def __init__(self, target, args=()):
        super().__init__()
        self._target = target
        self._args = args
        self._return = None
        self._exc = None

    def run(self):
        try:
            self._return = self._target(*self._args)
        except Exception as e:
            self._exc = e

    def join(self, timeout=None):
        super().join(timeout)
        if self.is_alive():
            raise TimeoutException("Timed out!")
        if self._exc:
            raise self._exc
        return self._return


# ----------------------------
# PDF processing
# ----------------------------
def pdf_to_pages(pdf_path):
    doc = fitz.open(pdf_path)
    pages = []
    for page in doc:
        text = page.get_text()
        # remove headers, footers, and page numbers
        text = re.sub(r"^\s*\d+\s+[A-Za-z ]+\s*$", "", text, flags=re.MULTILINE)
        # remove figure/table captions
        text = re.sub(r"(Figure|Table)\s+\d+(\.\d+)?[:.].*?(?=\n\n|\Z)", "", text, flags=re.DOTALL)
        # cleanup
        text = text.replace("\n", " ")
        text = re.sub(r"\s+", " ", text).strip()
        pages.append(text)
    return pages


# ----------------------------
# Audiobook rendering
# ----------------------------
def pdf_to_audiobook(pdf_path, out_path, model="tts_models/en/vctk/vits",
                     page_range=None, speaker="p274", timeout_factor=25):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    tts = TTS(model).to(device)

    pages = pdf_to_pages(pdf_path)
    if page_range:
        start, end = page_range
        pages = pages[start-1:end]  # 1-based input, Python is 0-based

    audio_streams = []
    for i, text in enumerate(pages, 1):
        print(f"Rendering page {i}/{len(pages)} on {device}...")

        # Split text into manageable chunks
        chunks = [text[j:j+2000] for j in range(0, len(text), 2000)]
        for chunk in chunks:
            expected_time = len(chunk) / 50  # heuristic ~50 chars/s
            max_time = int(timeout_factor * expected_time)

            thread = ThreadWithReturn(target=tts.tts, args=(chunk, speaker))
            try:
                thread.start()
                wav = thread.join(timeout=max_time)
                audio_streams.append(wav)
                print(f" > Done (limit {max_time}s)")
            except TimeoutException:
                print(f" > Skipping chunk (timeout {max_time}s)")

    if audio_streams:
        full_audio = np.concatenate(audio_streams)
        sf.write(out_path, full_audio, samplerate=tts.synthesizer.output_sample_rate)
        print(f"Saved audiobook: {out_path}")
    else:
        print("No audio generated.")


# ----------------------------
# Main
# ----------------------------
def main():
    pdf_to_audiobook(
        pdf_path="Satellite_Comms.pdf",
        model="tts_models/en/vctk/vits",
        out_path="Chapter_12.wav",
        page_range=(749, 764),  # <-- adjust to chapter/page range
        speaker="p274"
    )

if __name__ == "__main__":
    main()
```

```python
import os
import subprocess
from mutagen.easyid3 import EasyID3

# Settings
input_dir = ""  # folder containing Chapter_1.wav ... Chapter_13.wav
output_dir = "Satellite Communications"
audiobook_title = "Satellite Communications"
author = "Boppo"
year = "2025"

os.makedirs(output_dir, exist_ok=True)

# --- Pass 1: Encode with ffmpeg ---
for i in range(1, 14):
    in_file = os.path.join(input_dir, f"Chapter_{i}.wav")
    out_file = os.path.join(output_dir, f"Chapter_{i}.mp3")

    cmd = [
        "ffmpeg", "-y", "-i", in_file,
        "-codec:a", "libmp3lame", "-b:a", "192k",
        "-threads", "0", out_file
    ]
    subprocess.run(cmd, check=True)
    print(f"Encoded {in_file} → {out_file}")

# --- Pass 2: Tag with mutagen ---
for i in range(1, 14):
    out_file = os.path.join(output_dir, f"Chapter_{i}.mp3")
    tags = EasyID3(out_file)
    tags["title"] = f"Chapter {i}"
    tags["artist"] = author
    tags["album"] = audiobook_title
    tags["tracknumber"] = str(i)
    tags["date"] = year
    tags["genre"] = "Audiobook"
    tags.save()
    print(f"Tagged {out_file}")

print("All chapters encoded and tagged.")
```
